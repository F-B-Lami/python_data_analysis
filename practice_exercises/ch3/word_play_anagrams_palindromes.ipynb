## anagram

```python
# load word.txt
# brute method
words = []
file_words = open("words.txt", "r")
for line in file_words:
    words.append(line)

words[:15]
```

```python
# comprehension
words = [line for line in open('words.txt', 'r')]
num_total = len(words)
print(num_total)
words[:15]
```

```python
# comprehension with strip for spaces (leading and trailing + \n) using sets (auto remove duplicates)
words = sorted({line.strip().lower() for line in open('words.txt', 'r')})
print('num_of_duplicates = ', (num_total - len(words)))
words[:15]
```

```python
# function signature
def sig(word):
    return ''.join(sorted(word))
```

```python
sig('lami')
```

```python
# brute force for single anagram
def find_anagram_brute_force(word):
    try:
        return [word_check for word_check in words if sig(word_check) == sig(word)]
    except:
        return []
```

```python
%time
find_anagram_brute_force('aaru')
```

```python
# find all anagrams 
import collections
def all_anagrams():
    words_by_sig = collections.defaultdict(set)
    for word in words:
        words_by_sig[sig(word)].add(word)
    return words_by_sig
```

```python
all_anagrams_by_sig = {sig : anagrams for sig, anagrams in all_anagrams().items() if len(anagrams) > 1}
all_anagrams_by_sig
```

```python
def find_anagram_by_sig(sig_of_word):
    try:
        return all_anagrams_by_sig[sig_of_word]
    except:
        return []
```

```python
find_anagram_by_sig('aabelno')
```

```python
[all_anagrams_by_sig[sig_of_word] for sig_of_word in sorted(all_anagrams_by_sig.keys(), key=len, reverse=True)]
```

```python
[words for words in sorted(all_anagrams_by_sig.values(), key=len, reverse=True)]
```

## palindromes

```python
# load unique words, sorted
words = sorted({line.strip().lower() for line in open('words.txt', 'r')})
```

```python
import collections
def all_self_palindromes():
    words_by_sig = collections.defaultdict(set)
    for word in words:
        key = word[::-1]
        if key == word:
            words_by_sig[key].add(word)
    return words_by_sig

all_self_pals = {word : pals for word, pals in all_self_palindromes().items()}
print(len(list(all_self_pals.keys())))
all_self_pals
```

```python
# lesson solution using itertools
import itertools
list(itertools.combinations((1, 5, 7, 9), 3))
```

```python
# all pals are anagrams
pairs_pals = []
for wordset in all_anagrams_by_sig.values():
    for word1, word2 in itertools.combinations(wordset, 2):
        if word1[::-1] == word2:
            pairs_pals.append([word1, word2])
            
print(len(pairs_pals))
pairs_pals
```

```python
# elegant comprenhension
pair_palindromes = [[word1, word2] for wordset in all_anagrams_by_sig.values() for word1, word2 in itertools.combinations(wordset, 2) if word1[::-1] == word2]
print(len(pair_palindromes))
pair_palindromes
```
